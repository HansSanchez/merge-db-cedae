```python
# Importación de librerias
import pandas as pd

# Definición de la clase
class ImportClass:
    
    # Función inicial o constructor
    def __init__(self):
        # lectura de los archivos
        
## PASO 1: Lectura del archivo estadística_idc_2011_2015_2019 y separación en Dataframes por año
##------------------------------------------------------------------------------------------------------------------------##
        self.df_estadistica_idc_2011 = pd.read_excel(r'estadística_idc_2011_2015_2019.xlsx', 
                                                    sheet_name='2011', dtype={
                                                                               'dd': str,
                                                                               'mm': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Total': int
                                                                           }) 
        self.df_estadistica_idc_2015 = pd.read_excel(r'estadística_idc_2011_2015_2019.xlsx', 
                                                    sheet_name='2015', dtype={
                                                                               'dd': str,
                                                                               'mm': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Total': int
                                                                           }) 
        self.df_estadistica_idc_2019 = pd.read_excel(r'estadística_idc_2011_2015_2019.xlsx', 
                                                    sheet_name='2019', dtype={
                                                                               'dd': str,
                                                                               'mm': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Total': int
                                                                           })
##------------------------------------------------------------------------------------------------------------------------##
## PASO 2: Lectura del archivo delitos_electorales_spoa_2011_2015_2019
        self.df_delitos_electorales_spoa_2011_2015_2019 = pd.read_csv(r'delitos_electorales_spoa_2011_2015_2019.csv', sep=';', encoding='UTF-8-SIG') 
```

```python
# Instancia del objeto
IC = ImportClass()
```

```python
# Definición de variables por año
df_estadistica_idc_2011 = IC.df_estadistica_idc_2011
df_estadistica_idc_2015 = IC.df_estadistica_idc_2015
df_estadistica_idc_2019 = IC.df_estadistica_idc_2019
df_delitos_electorales_spoa_2011_2015_2019 = IC.df_delitos_electorales_spoa_2011_2015_2019
```

```python
def replace_characters(text):
      dictionary_replacement = {
          '√Å': 'Á',
          '√°': 'Á',
          '√â': 'É',
          '√©': 'É',
          '√Â': 'É',
          '√≠': 'Í',
          '√Ç': 'Í',
          '√Ì': 'Ó',
          '√≥': 'Ó',
          '√ì': 'Ó',
          '√ö': 'Ú',
          '√Ö': 'Ú',
          '√±': 'Ñ',
          '√Ë': 'Ñ',
      }
    
      for char, replacement in dictionary_replacement.items():
            text = text.replace(char, replacement)
    
      return text
```

```python
# Lista de columnas a las que se aplicará la función
columns = ['DELITO', 'CRIMINALIDAD', 'ES_ARCHIVO', 'ES_PRECLUSI√ìN', 'ESTADO', 'ETAPA_CASO', 'LEY', 'DEPARTAMENTO_HECHO', 'MUNICIPIO_HECHO']

# Aplicar la función a las columnas
df_delitos_electorales_spoa_2011_2015_2019[columns] = df_delitos_electorales_spoa_2011_2015_2019[columns].applymap(replace_characters)
```
