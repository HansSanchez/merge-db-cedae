```python
# Importación de librerias
import pandas as pd

# Definición de la clase
class ImportClass:
    
    # Función inicial o constructor
    def __init__(self):
        # lectura de los archivos
        
## PASO 1: Lectura del archivo estadística_idc_2011_2015_2019 y separación en Dataframes por año
##------------------------------------------------------------------------------------------------------------------------##
        self.df_estadistica_idc_2011 = pd.read_excel(r'estadística_idc_2011_2015_2019.xlsx', 
                                                    sheet_name='2011', dtype={
                                                                               'dd': str,
                                                                               'mm': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Total': int
                                                                           }) 
        self.df_estadistica_idc_2015 = pd.read_excel(r'estadística_idc_2011_2015_2019.xlsx', 
                                                    sheet_name='2015', dtype={
                                                                               'dd': str,
                                                                               'mm': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Total': int
                                                                           }) 
        self.df_estadistica_idc_2019 = pd.read_excel(r'estadística_idc_2011_2015_2019.xlsx', 
                                                    sheet_name='2019', dtype={
                                                                               'dd': str,
                                                                               'mm': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Total': int
                                                                           })
##------------------------------------------------------------------------------------------------------------------------##
## PASO 2: Lectura del archivo delitos_electorales_spoa_2011_2015_2019
        self.df_delitos_electorales_spoa_2011_2015_2019 = pd.read_csv(r'delitos_electorales_spoa_2011_2015_2019.csv', sep=';', encoding='UTF-8-SIG') 
```

```python
def replace_characters(text):
#       dictionary_replacement = {
#           '√Å': 'Á',
#           '√°': 'Á',
#           '√â': 'É',
#           '√©': 'É',
#           '√Â': 'É',
#           '√≠': 'Í',
#           '√Ç': 'Í',
#           '√Ì': 'Ó',
#           '√≥': 'Ó',
#           '√ì': 'Ó',
#           '√ö': 'Ú',
#           '√Ö': 'Ú',
#           '√±': 'Ñ',
#           '√Ë': 'Ñ',
#       }

      dictionary_replacement = {
          '√Å': 'A',
          '√°': 'A',
          '√â': 'E',
          '√©': 'E',
          '√Â': 'E',
          '√≠': 'I',
          '√Ç': 'I',
          '√Ì': 'O',
          '√≥': 'O',
          '√ì': 'O',
          '√ö': 'U',
          '√Ö': 'U',
          '√±': 'Ñ',
          '√Ë': 'Ñ',
      }
    
      for char, replacement in dictionary_replacement.items():
            text = text.replace(char, replacement)
    
      return text
```

```python
# Lista de columnas a las que se aplicará la función
columns = ['DELITO', 'CRIMINALIDAD', 'ES_ARCHIVO', 'ES_PRECLUSI√ìN', 'ESTADO', 'ETAPA_CASO', 'LEY', 'DEPARTAMENTO_HECHO', 'MUNICIPIO_HECHO']

# Aplicar la función a las columnas
df_delitos_electorales_spoa_2011_2015_2019[columns] = df_delitos_electorales_spoa_2011_2015_2019[columns].applymap(replace_characters)
```

```python
df_delitos_electorales_spoa_2011Copy = df_delitos_electorales_spoa_2011.reset_index().drop(['index'], axis=1).copy()
```

```python
# Crear un diccionario de mapeo de departamentos y valores a partir de df_estadistica_idc_2011
departamentos_valores = dict(zip(df_estadistica_idc_2011['departament'], df_estadistica_idc_2011['cod_dep']))

# Define una función para aplicar la condición utilizando el diccionario
def asignar_valor(row):
    departamento = row['departament']
    return departamentos_valores.get(departamento, None)  # Devuelve el valor correspondiente o None si no se encuentra

# Aplica la función a cada fila del DataFrame y crea la nueva columna
df_delitos_electorales_spoa_2011Copy['cod_dep'] = df_delitos_electorales_spoa_2011Copy.apply(asignar_valor, axis=1)
```

```python
# Crear un diccionario de mapeo de departamentos y valores a partir de df_estadistica_idc_2011
municipality_valores = dict(zip(df_estadistica_idc_2011['municipality'], df_estadistica_idc_2011['cod_mun']))

# Define una función para aplicar la condición utilizando el diccionario
def asignar_valor(row):
    municipality = row['municipality']
    return municipality_valores.get(municipality, None)  # Devuelve el valor correspondiente o None si no se encuentra

# Aplica la función a cada fila del DataFrame y crea la nueva columna
df_delitos_electorales_spoa_2011Copy['cod_mun'] = df_delitos_electorales_spoa_2011Copy.apply(asignar_valor, axis=1)
```
