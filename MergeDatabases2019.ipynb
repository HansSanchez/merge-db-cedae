```python

# Importación de librerias
import pandas as pd

# Definición de la clase
class ImportClass:
    
    # Función inicial o constructor
    def __init__(self):
        # lectura de los archivos
        
## PASO 1: Lectura del archivo estadística_idc_2011_2015_2019 y separación en Dataframes por año
##------------------------------------------------------------------------------------------------------------------------##
        self.df_estadistica_idc_2019 = pd.read_excel(r'estadística_idc_2011_2015_2019.xlsx', 
                                                    sheet_name='2019', dtype={
                                                                               'dd': str,
                                                                               'mm': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Total': int
                                                                           }) 
##------------------------------------------------------------------------------------------------------------------------##
## PASO 2: Lectura del archivo delitos_electorales_spoa_2011_2015_2019
        self.df_delitos_electorales_spoa_2011_2015_2019 = pd.read_csv(r'delitos_electorales_spoa_2011_2015_2019.csv', sep=';', encoding='UTF-8-SIG') 
    
##------------------------------------------------------------------------------------------------------------------------##
## PASO 3: Lectura del archivo trashumancia_2019
        self.df_trashumancia_2019 = pd.read_excel(r'trashumancia_2019.xlsx', 
                                                    sheet_name='Trash2019', dtype={
                                                                               'Orden': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Trashumantes': int
                                                                           })
##------------------------------------------------------------------------------------------------------------------------##
## PASO 4: Lectura del archivo divipole_autoridades_locales_2019
        self.df_divipole_autoridades_territoriales_2019 = pd.read_excel(r'divipole_autoridades_territoriales_2019.xlsx', 
                                                    sheet_name='Base', dtype={
                                                                               'dd': str,
                                                                               'mm': str,
                                                                               'zz': str,
                                                                               'pp': str,
                                                                               'Departamento': str,
                                                                               'Municipio': str,
                                                                               'Puesto': str,
                                                                               'Mujeres': int,
                                                                               'Hombres': int,
                                                                               'Total': int,
                                                                               'Mesas': int,
                                                                               'Dirección': str
                                                                           })    

```

```python
def replace_characters(text):
#       dictionary_replacement = {
#           '√Å': 'Á',
#           '√°': 'Á',
#           '√â': 'É',
#           '√©': 'É',
#           '√Â': 'É',
#           '√≠': 'Í',
#           '√Ç': 'Í',
#           '√Ì': 'Ó',
#           '√≥': 'Ó',
#           '√ì': 'Ó',
#           '√ö': 'Ú',
#           '√Ö': 'Ú',
#           '√±': 'Ñ',
#           '√Ë': 'Ñ',
#       }

      dictionary_replacement = {
          '√Å': 'A',
          '√°': 'A',
          '√â': 'E',
          '√©': 'E',
          '√Â': 'E',
          '√≠': 'I',
          '√Ç': 'I',
          '√Ì': 'O',
          '√≥': 'O',
          '√ì': 'O',
          '√ö': 'U',
          '√Ö': 'U',
          '√±': 'Ñ',
          '√Ë': 'Ñ',
      }
    
      for char, replacement in dictionary_replacement.items():
            text = text.replace(char, replacement)
    
      return text
```

```python
# Lista de columnas a las que se aplicará la función
columns = ['DELITO', 'CRIMINALIDAD', 'ES_ARCHIVO', 'ES_PRECLUSI√ìN', 'ESTADO', 'ETAPA_CASO', 'LEY', 'DEPARTAMENTO_HECHO', 'MUNICIPIO_HECHO']

# Aplicar la función a las columnas
df_delitos_electorales_spoa_2011_2015_2019[columns] = df_delitos_electorales_spoa_2011_2015_2019[columns].applymap(replace_characters)
```

```python
# Crear un DataFrame vacío para almacenar los resultados
df_consolidated_2019 = pd.DataFrame(columns=df_dep_mun.columns)

# Iterar a través de las filas de df_dep_mun
for _, fila in df_dep_mun.iterrows():
    municipio = fila['municipality']
    # Buscar el municipio correspondiente en df_2019
    filtro = df_2019['municipality'] == municipio
    datos_municipio = df_2019[filtro]
    
    if not datos_municipio.empty:
        # Si se encuentra el municipio en df_2019, agregamos las filas correspondientes
        df_consolidated_2019 = pd.concat([df_consolidated_2019, datos_municipio], ignore_index=True)
    else:
        # Si no se encuentra el municipio en df_2019, agregamos una fila vacía
        df_consolidated_2019 = pd.concat([df_consolidated_2019, fila.to_frame().T], ignore_index=True)

# Ahora, df_consolidated_2019 contiene los datos combinados
```

```python
df_consolidated_2019
```

```python
df_consolidated_2019['year'] = '2019'
columnas_para_reemplazar = ['crime', 'criminality', 'is_file', 'is_preclusion', 'state', 'state_case', 'law']

for columna in columnas_para_reemplazar:
    df_consolidated_2019[columna] = df_consolidated_2019[columna].fillna('SIN REGISTRO')
```

```python
columnas_para_reemplazar_int = ['total_processes', 'transhumants', 'total_idc', 'total_women', 'total_men', 'total_people']

for columna in columnas_para_reemplazar_int:
    df_consolidated_2019[columna] = df_consolidated_2019[columna].fillna(0)
```
